# Importing libraries and loading csvs as dataframes

```python
# creating and using a Boolean mask to select data by date
prePilotMask = (df_turn_restrict1718['dt_hourly'] >= '2017-10-03') & (df_turn_restrict1718['dt_hourly'] < '2017-11-12')
df_prePilot = df_turn_restrict1718.loc[prePilotMask].copy()
df_prePilot["day"] = df_prePilot["dt_hourly"].dt.day
df_prePilot["hour"] = df_prePilot["dt_hourly"].dt.hour
df_prePilot
```

# Grouping by datetime and plotting with matplotlib

## Datecount group and plot

```python
df_prePilot["date"] = df_prePilot["dt_hourly"].apply(lambda x: x.date())

df_pp_datecount = df_prePilot.groupby(["date"])["intersection_uid"].value_counts()

unstack_pp_datecount = df_pp_datecount.unstack()
unstack_pp_datecount.plot(kind="bar", stacked=True, title="records per day by intersection")
plt.legend(bbox_to_anchor=(1, 1), loc='upper left')
plt.show()
```

UID 10 started on 10-03

UID 20 started on 10-04

as the intersections at the beginning and end of the corridor, they appear to have the most consistent data

11-07, something must of happened, data collection starts at 08:00 instead of 00:00

absolutely no data for Oct 16 to 29 for wahtever reason

## Hourcount group and plot

```python
df_pp_hourcount = df_prePilot.groupby(["hour"])["intersection_uid"].value_counts()

unstack_pp_hourcount = df_pp_hourcount.unstack()
unstack_pp_hourcount.plot(kind="bar", stacked=True, title="records per hour by intersection")
plt.legend(bbox_to_anchor=(1, 1), loc='upper left')
plt.show()
```

for this pre-pilot period, there were fewer records collected between 23:00 and 7:00

# (post) Pilot group and plot

```python
# creating and using a Boolean mask to select data by date
PilotMask = (df_turn_restrict1718['dt_hourly'] >= '2017-11-12') & (df_turn_restrict1718['dt_hourly'] < '2018-08-25')
df_Pilot = df_turn_restrict1718.loc[PilotMask].copy()
df_Pilot["date"] = df_Pilot["dt_hourly"].apply(lambda x: x.date())
df_Pilot["hour"] = df_Pilot["dt_hourly"].apply(lambda x: x.hour)

df_datecount = df_Pilot.groupby(["date"])["intersection_uid"].value_counts()
df_hourcount = df_Pilot.groupby(["hour"])["intersection_uid"].value_counts()
```

```python
unstack_datecount = df_datecount.unstack()
unstack_datecount.plot(kind="bar", stacked=True, title="records per day by intersection")
plt.legend(bbox_to_anchor=(1, 1), loc='upper left')
plt.show()

unstack_hourcount = df_hourcount.unstack()
unstack_hourcount.plot(kind="bar", stacked=True, title="records per hour by intersection")
plt.legend(bbox_to_anchor=(1, 1), loc='upper left')
plt.show()
```

post-pilot data until 2018, lots of dates missing, no data until a month later (December)

UID 14 dropped from reporting

hours of reporting vary from 00-23, 06-21, 06-23, 00-03 + 09-23, 00-03 + 08-23, 00-03 + 08-21

concentration of records are 06-21, with special interest in 8am/9am and 4/5/6pm
